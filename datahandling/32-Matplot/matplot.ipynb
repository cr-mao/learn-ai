通过 fig,(ax1,ax2)=plt.subplots(2,1) 这句代码，我们定义了一个 Figure 画布，同时利用 subplots(2,1) 函数将画布分为 2 行 1 列的两部分，其中一部分命名为 ax1，另外一部分为 ax2。这样通过上述操作，我们就能够在一个画布中绘制多个图表。

```python
import matplotlib.pyplot as plt
fig,(ax1,ax2)=plt.subplots(2,1) # 划分区域
ax1.plot()
ax2.plot()
plt.show()
# import matplotlib
# print(matplotlib.matplotlib_fname())
```

add_subplot(2,1,1)函数包括三个参数，前两个指出如何对 Figure 画布进行划分，后面的参数指出选中第几部分

```python
import matplotlib.pyplot as plt
fig=plt.figure() # 创建画纸
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
data=pd.read_excel('fight.xlsx')
x=data.派遣人数
y=data.攻占用时
plt.scatter(x,y)
plt.title('攻占时间和进攻人数')
plt.ylabel('攻占用时')
plt.xlabel('进攻人数')
plt.gca().yaxis.set_major_formatter(mticker.FormatStrFormatter('%.0fs'))
plt.show()
```

### 气泡图
气泡图其实就是散点图的升级，通过控制点的大小来展现新的数据。

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
plt.rcParams['font.sans-serif']='SimHei'
plt.rcParams['axes.unicode_minus']=False
data=pd.read_excel('fights.xlsx')
x=data.派遣人数
y=data.攻占用时
s=data.坦克数量
# alpha透明度，s=size 
plt.scatter(x,y,s=s*200+20,alpha=0.5)
plt.title('攻占时间和进攻人数')
plt.ylabel('攻占用时')
plt.xlabel('进攻人数')
plt.gca().yaxis.set_major_formatter(mticker.FormatStrFormatter('%.0fs'))
plt.show()
```
