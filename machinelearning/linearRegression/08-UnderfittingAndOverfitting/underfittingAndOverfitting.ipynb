## 过拟合与欠拟合

### 通过代码认识过拟合和欠拟合

### 欠拟合示例

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(666)
#-3,3 随机概率 取１００个元素
x = np.random.uniform(-3,3,size = 100)
# 线性回归模型需要二维数组
X = x.reshape(-1,1)
# 抛物线 ，np.random.normal(0,1,size = 100)正太分布为了增加噪声，所以变成非抛物线
y = 0.5* x**2 + x+2 +np.random.normal(0,1,size = 100)

from sklearn.linear_model import LinearRegression
estimator = LinearRegression()
estimator.fit(X,y)
y_predict = estimator.predict(X)

plt.scatter(x,y)
plt.plot(x,y_predict,color = 'r')
plt.show()

```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y,y_predict)
```

### 正好拟合



```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(666)
#-3,3 随机概率 取１００个元素
x = np.random.uniform(-3,3,size = 100)
# 线性回归模型需要二维数组
X = x.reshape(-1,1)
# 抛物线 ，np.random.normal(0,1,size = 100)正太分布为了增加噪声，所以变成非抛物线
y = 0.5* x**2 + x+2 +np.random.normal(0,1,size = 100)

# 线性回归模型需要二维数组
X = x.reshape(-1,1)
# 增加 一个平方特征 ，变２个列了
X2 = np.hstack([X,X**2])
estimator2 = LinearRegression()
estimator2.fit(X2,y)
y_predict2 = estimator2.predict(X2)

plt.scatter(x,y)
# 从小排序， plot折线图
plt.plot(np.sort(x),y_predict2[np.argsort(x)],color = 'r')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y,y_predict2)
```

### 一起写

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(666)
#-3,3 随机概率 取１００个元素
x = np.random.uniform(-3,3,size = 100)
# 线性回归模型需要二维数组
X = x.reshape(-1,1)
# 抛物线 ，np.random.normal(0,1,size = 100)正太分布为了增加噪声，所以变成非抛物线
y = 0.5* x**2 + x+2 +np.random.normal(0,1,size = 100)

# 线性回归模型需要二维数组
X = x.reshape(-1,1)
# 增加 一个平方特征 ，变２个列了
X2 = np.hstack([X,X**2])
#特征过多
X5 = np.hstack([X2,X**3,X**4,X**5,X**6,X**7,X**8,X**9,X**10])

# 欠拟合
estimator= LinearRegression()
estimator.fit(X,y)
# 正好拟合
estimator2 = LinearRegression()
estimator2.fit(X2,y)

estimator3 = LinearRegression()
estimator3.fit(X5,y)
y_predict5 = estimator3.predict(X5)

y_predict = estimator.predict(X)
y_predict2 = estimator2.predict(X2)

plt.scatter(x,y)
plt.plot(x,y_predict,color = 'g')
# 从小排序， plot折线图
plt.plot(np.sort(x),y_predict2[np.argsort(x)],color = 'r')

plt.plot(np.sort(x),y_predict5[np.argsort(x)],color = 'y')
plt.show()

error = mean_squared_error(y, y_predict)
print(error)

error = mean_squared_error(y, y_predict2)
print(error)
error = mean_squared_error(y, y_predict5)
print(error)

```
